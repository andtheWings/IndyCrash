```python
import numpy as np
import pandas as pd
import matplotlib
```

# Import Data

```python
# Obtain from https://hub.mph.in.gov/dataset/aries-crash-data-2007-2017/resource/911880b1-d0bb-4168-a5c1-a52dcc291445

raw_df = pd.read_csv('aries_crash_data_2017.csv')
raw_df.head()
```

# Filter Dataframe

```python
raw_df['AGE_GRP'].value_counts()
```

```python
raw_df['PERSONTYPEDESCR'].value_counts()
```

```python
young = (raw_df['AGE_GRP'] == "<1 Year") | (raw_df['AGE_GRP'] == "1-4 Years") | (raw_df['AGE_GRP'] == "5-14 Years") | (raw_df['AGE_GRP'] == "15-24 Years")
on_street = (raw_df['PERSONTYPEDESCR'] == "Pedestrian") | (raw_df['PERSONTYPEDESCR'] == "Pedal Cyclist")

df = raw_df[young & on_street]
df.head()
```

```python
c = pd.crosstab(df['AGE_GRP'], df['PERSONTYPEDESCR'], margins=True, margins_name="Total")
c = c.reindex(["<1 Year","1-4 Years","5-14 Years","15-24 Years","Total"], columns=["Pedestrian","Pedal Cyclist","Total"])
c
```
